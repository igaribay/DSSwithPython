## ```If``` Statements

The general python syntax for a ```if``` statement is

```if``` condition:

    statements 

````python
# odd or even

num = int (raw_input("Enter your number: "))

if num%2 == 0:
    print("Number is Even")

if num%2 != 0:
    print("Number is odd")
    
    
print("Done!")
````

````python
number = int(raw_input("Enter your number: "))

if number < 100:
    if number%2 == 0:
        if number%5 == 0:
            print (number)
        if number%5 != 0:
            print (number * 5)
            
    if number%2 != 0:
        print (number * 2)
        
if number >= 100:
    print ("The number is out of range!")
    
print("Done!")
            
        
    
````

## ```if - else``` Statements

By using this statement we can reduce codes in the previous example:

````python
number = int(raw_input("Enter your number: "))

if number < 100:
    if number%2 == 0:
        if number%5 == 0:
            print (number)
        else:
            print (number * 5)
            
    else:
        print (number * 2)
        
if number > 100:
    print ("The number is out of range!")
    
print("Done!")
            
````

````python
# The number of days in a month

days = int(raw_input("How many days are in February? "))

if days == 28:
    print("You got it.")
    
else:
    print("Try again!")
    
print("Thank you")
````

## Loops in Python

When you want to repeat a fixed number of times, loops are useful to reduce the number of iterations.

There are two types of loops in python: ```for``` and ```while```

```while``` condition:

    statements

````python
num = 7

while num > 0:
    print (num)
    num = num - 1
````

````python
Day = ["sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]

Count = 0

while Count <= 6:
    print ("Today is : ", Day[Count])
    Count = Count + 1
````

````python
num = int(raw_input("Enter your number : "))

while num < 10:
    while num <= 5:
        print (num * 4)
        num = num + 1
        
    else:
        print (num - 1)
        num = num + 1
        
else:
    print (num // 3)
    num = num + 1
    
print("Done!")
````

````python
Word = str (raw_input("Enter your word: ")) # Use "raw_input" for Python2 and "input" for Python3

index = 0
List_Word = []

while index < len (Word):
    letter = Word[index]
    List_Word.append(letter)
    index = index + 1
    
print (List_Word)
````

```for``` variable ```in``` iterator:

    statements

````python
Word = str (raw_input("Enter your word: "))

for letter in range (len(Word)):
    print (Word[letter])
    
````

````python
for x in range (2,4):
    for y in range (3,6):
        print (x * y)
````

````python
# counting the number of specific letter in a sentence

Sentence = str(raw_input("Enter your sentence: "))

count = 0

for char in Sentence:
    if char == "a":
        count = count + 1
        
print(count)
````

Some more example:


````python
# using "break" in loop 

value = 7

while value > 0:
    print(value)
    value = value - 1
    if value == 4:
        break
        
print("End")
````

````python
# using "continue" in loop 

value = 7

while value > 0:
    value = value - 1
    if value == 4:
        continue
    print (value)
        
print("End")
````

````python
# Identify the prime number

for number in range (2,30):
    for x in range (2,number):
        if number % x == 0:
            print (number, "is not prime!")
            break               # check the result with/without break and compare them. 
    else:
        print (number, "is prime!" )
````

````python
for i in range (20,40):
    if 20 <= i <= 30:
        continue
    print (i)
````

### Introducing functions in python

If you want to define a function in python,you need to use the keyword ```def``` 

The structure of defining a function is like below:



```def``` functionName (V,a):

    statements

````python
# Is a number prime?

def prime (Candidate_num):
    if Candidate_num > 1:
        for x in range (2,Candidate_num):
            if Candidate_num % x == 0:
                return False
    return True


print (prime(13))
````

````python
Number = int (raw_input("Enter your number: "))

print (prime (Number))

````

````python
# Defining factorial function

def factorial (number):
    if number == 0:
        return 1
    else:
        return number * factorial (number - 1)
    
    
    
print (factorial (7))
````
